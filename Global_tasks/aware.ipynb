````python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# 1. Resmi oku
img = cv2.imread('fotoÄŸraf_yeni_kopya.png')

# 2. Resmi Ä°stenen Boyuta Getir (300 GeniÅŸlik, 300 YÃ¼kseklik)
# DÄ°KKAT: cv2.resize (geniÅŸlik, yÃ¼kseklik) alÄ±r.
img_resized = cv2.resize(img, (300, 300))

# 3. Ä°ÅŸlem YapÄ±lacak AlanÄ±n KoordinatlarÄ±
# Resim artÄ±k 330x300 olduÄŸu iÃ§in 100-200 aralÄ±ÄŸÄ± bu yeni boyuta gÃ¶re seÃ§ilecek.
y1, y2 = 100, 200
x1, x2 = 100, 200

# --- ADIM A: AlanÄ± SeÃ§ (ROI) ---
roi_alan = img_resized[y1:y2, x1:x2]

# --- ADIM B: Griye Ã‡evir ---
roi_gri = cv2.cvtColor(roi_alan, cv2.COLOR_BGR2GRAY)

# --- ADIM C: Tekrar 3 KanallÄ± (BGR) Formata Ã‡evir ---
# (Gri gÃ¶rÃ¼ntÃ¼yÃ¼ renkli resmin iÃ§ine geri yapÄ±ÅŸtÄ±rabilmek iÃ§in)
roi_gri_bgr = cv2.cvtColor(roi_gri, cv2.COLOR_GRAY2BGR)

# --- ADIM D: BulanÄ±klaÅŸtÄ±r ---
roi_bulanik = cv2.GaussianBlur(roi_gri_bgr, (15, 15), 0)

# --- ADIM E: Ä°ÅŸlenmiÅŸ ParÃ§ayÄ± Yerine Koy ---
img_resized[y1:y2, x1:x2] = roi_bulanik

# 4. SonuÃ§larÄ± GÃ¶ster ve Kaydet
print(f"330x300 BoyutlandÄ±rÄ±lmÄ±ÅŸ ve AlanÄ± Ä°ÅŸlenmiÅŸ Resim:")
cv2_imshow(img_resized)

cv2.imwrite('fotoÄŸraf_son_gorev.png', img_resized)
print("Ä°ÅŸlem tamamlandÄ±, 'fotoÄŸraf_son_gorev.png' kaydedildi.")
````

1. Thresholding (EÅŸikleme) Nedir?
Thresholding, gri tonlamalÄ± (grayscale) bir gÃ¶rÃ¼ntÃ¼yÃ¼ ikili (binary) gÃ¶rÃ¼ntÃ¼ye dÃ¶nÃ¼ÅŸtÃ¼rme iÅŸlemidir. Yani resimdeki pikselleri, belirli bir sÄ±nÄ±r deÄŸerine (eÅŸik) gÃ¶re sadece siyah (0) veya beyaz (255) olarak sÄ±nÄ±flandÄ±rmaktÄ±r.

MantÄ±ÄŸÄ± ÅŸudur:

Bir eÅŸik deÄŸeri (threshold value) belirlenir (Ã¶rneÄŸin 127).

Resimdeki her bir piksel tek tek kontrol edilir.

EÄŸer pikselin parlaklÄ±k deÄŸeri eÅŸik deÄŸerinden bÃ¼yÃ¼kse, o piksel beyaza atanÄ±r.

KÃ¼Ã§Ã¼kse, siyaha atanÄ±r.

Bu iÅŸlem, nesneleri arka plandan ayÄ±rmak (segmentasyon) iÃ§in en basit ve etkili yÃ¶ntemdir.

2. Hangi Alanlarda KullanÄ±lÄ±r?
Thresholding, gÃ¶rÃ¼ntÃ¼ iÅŸlemenin "Ã¶n iÅŸleme" aÅŸamasÄ± olarak pek Ã§ok sektÃ¶rde kritiktir:

Belge Tarama (OCR): TaranmÄ±ÅŸ belgelerdeki yazÄ±larÄ± kaÄŸÄ±t arka planÄ±ndan ayÄ±rmak iÃ§in kullanÄ±lÄ±r (Adobe Scan veya CamScanner gibi uygulamalar bunu yapar).

TÄ±bbi GÃ¶rÃ¼ntÃ¼leme: RÃ¶ntgen veya MR gÃ¶rÃ¼ntÃ¼lerinde kemik, tÃ¼mÃ¶r veya damar yapÄ±larÄ±nÄ± belirginleÅŸtirmek iÃ§in.

Robotik ve Otonom Sistemler: Robotun yerde takip edeceÄŸi ÅŸeridi (beyaz Ã§izgi) asfalttan ayÄ±rmak veya engelleri tespit etmek iÃ§in.

Kalite Kontrol: Ãœretim bandÄ±ndaki bir parÃ§anÄ±n Ã¼zerindeki Ã§atlaklarÄ± veya hatalarÄ± tespit etmek iÃ§in.

TarÄ±m Teknolojileri: Bitki yapraklarÄ±nÄ± topraktan ayÄ±rÄ±p hastalÄ±k tespiti yapmak veya bitki yoÄŸunluÄŸunu Ã¶lÃ§mek iÃ§in.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. GÃ¶rÃ¼ntÃ¼yÃ¼ Okuma
# 'yaprak.png' dosyasÄ±nÄ±n Colab'de "Files" (Dosyalar) sekmesinde olduÄŸundan emin olun.
# 0 parametresi resmi doÄŸrudan gri tonlamalÄ± (grayscale) okur.
img = cv2.imread('yaprak.png', 0)

if img is None:
    print("Hata: 'yaprak.png' dosyasÄ± bulunamadÄ±. LÃ¼tfen dosya adÄ±nÄ± kontrol edin veya dosyayÄ± yÃ¼kleyin.")
else:
    # 2. Thresholding Ä°ÅŸlemleri

    # A) Basit Thresholding
    # Piksel deÄŸeri 127'den kÃ¼Ã§Ã¼kse 0 (Siyah), bÃ¼yÃ¼kse 255 (Beyaz) yapar.
    # EÄŸer yaprak koyu, arka plan aÃ§Ä±ksa; yaprak siyah, arka plan beyaz olur.
    # (Tersini isterseniz cv2.THRESH_BINARY_INV kullanabilirsiniz)
    ret, thresh_simple = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

    # B) Otsu Thresholding (Otomatik EÅŸik)
    # EÅŸik deÄŸerini histograma gÃ¶re otomatik belirler.
    ret_otsu, thresh_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # C) Adaptive Thresholding (UyarlamalÄ±)
    # IÅŸÄ±k daÄŸÄ±lÄ±mÄ± dengesizse (gÃ¶lge vb.) en iyi sonucu bu verir.
    thresh_adaptive = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                            cv2.THRESH_BINARY, 11, 2)

    # 3. SonuÃ§larÄ± GÃ¶sterme
    titles = ['Orijinal Resim', 'Basit Threshold (127)', f'Otsu (Otomatik DeÄŸer: {ret_otsu})', 'Adaptive Threshold']
    images = [img, thresh_simple, thresh_otsu, thresh_adaptive]

    plt.figure(figsize=(15, 10))
    for i in range(4):
        plt.subplot(2, 2, i+1)
        plt.imshow(images[i], 'gray')
        plt.title(titles[i])
        plt.axis('off')

    plt.show()
````

````python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab import files # Dosya yÃ¼kleme kÃ¼tÃ¼phanesi
from google.colab.patches import cv2_imshow

# --- ADIM 1: MasaÃ¼stÃ¼nden Dosya YÃ¼kleme ---
print("LÃ¼tfen masaÃ¼stÃ¼nden kullanmak istediÄŸin resmi seÃ§:")
uploaded = files.upload()

# YÃ¼klenen dosyanÄ±n ismini otomatik alalÄ±m
if len(uploaded) > 0:
    filename = next(iter(uploaded)) # YÃ¼klenen ilk dosyanÄ±n adÄ±nÄ± al
    print(f"\nÄ°ÅŸlenecek dosya: {filename}")

    # Resmi oku
    img = cv2.imread(filename)

    if img is None:
        print("Hata: Resim okunamadÄ±.")
    else:
        # --- ADIM 2: Ã–n Ä°ÅŸlemler ---
        output_img = img.copy()
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Otsu eÅŸikleme
        ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

        # GÃ¼rÃ¼ltÃ¼ temizleme (Opening)
        kernel = np.ones((3,3), np.uint8)
        opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)

        # --- ADIM 3: Watershed HazÄ±rlÄ±ÄŸÄ± (Arka/Ã–n Plan) ---
        # Kesin arka plan (Sure Background)
        sure_bg = cv2.dilate(opening, kernel, iterations=3)

        # Kesin Ã¶n plan (Sure Foreground) - Distance Transform
        dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)

        # Birbirine deÄŸen nesneleri ayÄ±rmak iÃ§in eÅŸikleme
        # 0.5 oranÄ±nÄ± duruma gÃ¶re deÄŸiÅŸtirebilirsin
        ret, sure_fg = cv2.threshold(dist_transform, 0.5 * dist_transform.max(), 255, 0)

        # Bilinmeyen bÃ¶lge (Unknown)
        sure_fg = np.uint8(sure_fg)
        unknown = cv2.subtract(sure_bg, sure_fg)

        # --- ADIM 4: Etiketleme ve Watershed ---
        ret, markers = cv2.connectedComponents(sure_fg)
        markers = markers + 1 # Arka planÄ± 0'dan 1'e Ã§ek
        markers[unknown == 255] = 0 # Bilinmeyenleri 0 yap

        # AlgoritmayÄ± Ã§alÄ±ÅŸtÄ±r
        markers = cv2.watershed(img, markers)

        # SÄ±nÄ±rlarÄ± Ã§iz (KÄ±rmÄ±zÄ±)
        output_img[markers == -1] = [0, 0, 255]

        # --- SONUÃ‡LARI GÃ–STER ---
        print("\n--- SONUÃ‡LAR ---")
        print("1. Distance Transform (AyÄ±rma MantÄ±ÄŸÄ±):")
        dist_display = cv2.normalize(dist_transform, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
        cv2_imshow(dist_display)

        print("\n2. Watershed Sonucu (KÄ±rmÄ±zÄ± Ã‡izgiler):")
        cv2_imshow(output_img)

else:
    print("HiÃ§bir dosya yÃ¼klemediniz.")
````

````python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt
import io

# 1. Bilgisayardan Dosya YÃ¼kleme
print("LÃ¼tfen masaÃ¼stÃ¼nÃ¼zdeki 'masa.jpeg' dosyasÄ±nÄ± seÃ§in:")
uploaded = files.upload()

# YÃ¼klenen dosyanÄ±n ismini alalÄ±m (sizin dosya isminiz neyse onu kullanÄ±r)
filename = next(iter(uploaded))

# 2. GÃ¶rÃ¼ntÃ¼yÃ¼ HafÄ±zadan Okuma
# DosyayÄ± diske kaydetmeden doÄŸrudan hafÄ±zadan okumak iÃ§in decode iÅŸlemi
file_bytes = np.frombuffer(uploaded[filename], dtype=np.uint8)
img = cv2.imdecode(file_bytes, cv2.IMREAD_COLOR)

# ... (Geri kalan iÅŸlemler aynÄ±) ...
if img is None:
    print("GÃ¶rÃ¼ntÃ¼ okunamadÄ±.")
else:
    # BGR -> RGB ve Gri DÃ¶nÃ¼ÅŸÃ¼m
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # --- Edge Detection (Canny) ---
    edges = cv2.Canny(gray, 100, 200)

    # --- Corner Detection (Harris) ---
    gray_float = np.float32(gray)
    dst = cv2.cornerHarris(gray_float, 2, 3, 0.04)
    dst = cv2.dilate(dst, None)

    img_corners = img_rgb.copy()
    img_corners[dst > 0.01 * dst.max()] = [255, 0, 0]

    # --- GÃ¶rselleÅŸtirme ---
    plt.figure(figsize=(18, 6))

    plt.subplot(1, 3, 1)
    plt.imshow(img_rgb)
    plt.title('Orijinal')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(edges, cmap='gray')
    plt.title('Kenar Tespiti (Edges)')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(img_corners)
    plt.title('KÃ¶ÅŸe Tespiti (Corners)')
    plt.axis('off')

    plt.show()
````

HSV Nedir? (Hue, Saturation, Value)
Bizim gÃ¶zÃ¼mÃ¼z ve ekranlar renkleri RGB (KÄ±rmÄ±zÄ±, YeÅŸil, Mavi) karÄ±ÅŸÄ±mÄ± olarak gÃ¶rÃ¼r. Ancak bilgisayarÄ±n bir nesneyi tanÄ±masÄ± iÃ§in RGB kÃ¶tÃ¼dÃ¼r Ã§Ã¼nkÃ¼ Ä±ÅŸÄ±k deÄŸiÅŸiminden Ã§ok etkilenir. (GÃ¶lgedeki kÄ±rmÄ±zÄ± bir elma, bilgisayar iÃ§in siyahtÄ±r).

HSV ise rengi insanlarÄ±n algÄ±ladÄ±ÄŸÄ± gibi tanÄ±mlar:

H - Hue (Renk Ã–zÃ¼): Rengin kendisidir (KÄ±rmÄ±zÄ±, SarÄ±, Mavi vb.). IÅŸÄ±k deÄŸiÅŸse de bu deÄŸer sabit kalÄ±r.

OpenCV'de AralÄ±ÄŸÄ±: 0 - 179 (Derece/2)

S - Saturation (Doygunluk): Rengin canlÄ±lÄ±ÄŸÄ±dÄ±r. DÃ¼ÅŸÃ¼kse grileÅŸir, yÃ¼ksekse parlar.

OpenCV'de AralÄ±ÄŸÄ±: 0 - 255

V - Value (ParlaklÄ±k): Rengin ne kadar aydÄ±nlÄ±k olduÄŸudur. KaranlÄ±k mÄ±, Ä±ÅŸÄ±klÄ± mÄ±?

OpenCV'de AralÄ±ÄŸÄ±: 0 - 255

Neden KullanÄ±yoruz? Siyah, Mavi veya SarÄ± bir nesneyi takip ederken RGB kullanÄ±rsak, odaya gÃ¼neÅŸ girdiÄŸinde kod bozulur. HSV kullanÄ±rsak, sadece "Hue" deÄŸerine odaklanÄ±rÄ±z ve Ä±ÅŸÄ±k deÄŸiÅŸimlerini (Value) gÃ¶rmezden gelebiliriz.

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Javascript, clear_output
from google.colab.output import eval_js
from base64 import b64decode
import ipywidgets as widgets

# 1. FotoÄŸraf Ã‡ekme Fonksiyonu
def take_photo():
  js = Javascript('''
    async function takePhoto() {
      const div = document.createElement('div');
      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});
      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();
      // KullanÄ±cÄ±dan bir tuÅŸa basmasÄ±nÄ± beklemiyoruz, hemen alÄ±yoruz
      await new Promise(resolve => setTimeout(resolve, 1000));
      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', 0.8);
    }
    ''')
  display(js)
  data = eval_js('takePhoto()')
  binary = b64decode(data.split(',')[1])
  with open('test_img.jpg', 'wb') as f:
    f.write(binary)
  return 'test_img.jpg'

print("Kameradan Ã¶rnek gÃ¶rÃ¼ntÃ¼ alÄ±nÄ±yor...")
filename = take_photo()
original_img = cv2.imread(filename)
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB) # Matplotlib iÃ§in RGB
hsv_img = cv2.cvtColor(original_img, cv2.COLOR_RGB2HSV) # Ä°ÅŸlem iÃ§in HSV

# 2. Ä°nteraktif GÃ¼ncelleme Fonksiyonu
def update_mask(h_min, h_max, s_min, s_max, v_min, v_max):
    # Alt ve Ãœst SÄ±nÄ±rlarÄ± Belirle
    lower = np.array([h_min, s_min, v_min])
    upper = np.array([h_max, s_max, v_max])

    # Maske OluÅŸtur
    mask = cv2.inRange(hsv_img, lower, upper)

    # Maskeyi orijinal resme uygula (Bitwise AND)
    # Sadece seÃ§ilen renkleri gÃ¶sterir, gerisini siyah yapar
    result = cv2.bitwise_and(original_img, original_img, mask=mask)

    # GÃ¶rselleÅŸtirme
    plt.figure(figsize=(14, 6))

    # Orijinal
    plt.subplot(1, 2, 1)
    plt.imshow(original_img)
    plt.title("Orijinal Resim")
    plt.axis('off')

    # FiltrelenmiÅŸ SonuÃ§
    plt.subplot(1, 2, 2)
    plt.imshow(result)
    plt.title(f"HSV Filtresi Sonucu\nHue: {h_min}-{h_max}")
    plt.axis('off')

    plt.show()

# 3. Trackbar'larÄ± (SÃ¼rgÃ¼leri) OluÅŸtur
print("\n--- HSV AYAR PANELÄ° ---")
print("SÃ¼rgÃ¼leri kaydÄ±rarak istediÄŸiniz rengi izole etmeye Ã§alÄ±ÅŸÄ±n.")

widgets.interact(update_mask,
                 h_min=widgets.IntSlider(min=0, max=179, step=1, value=0, description='H-Min'),
                 h_max=widgets.IntSlider(min=0, max=179, step=1, value=179, description='H-Max'),
                 s_min=widgets.IntSlider(min=0, max=255, step=1, value=50, description='S-Min'),
                 s_max=widgets.IntSlider(min=0, max=255, step=1, value=255, description='S-Max'),
                 v_min=widgets.IntSlider(min=0, max=255, step=1, value=50, description='V-Min'),
                 v_max=widgets.IntSlider(min=0, max=255, step=1, value=255, description='V-Max'));
````

````python
import torch
import cv2
import numpy as np
import base64
from google.colab.output import eval_js
from IPython.display import display, Javascript
import time

# 1. MODELÄ° HAZIRLA
print("Model hazÄ±rlanÄ±yor...")
# Standart model (yolov5s) - 80 nesne
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', force_reload=False)
model.conf = 0.35
print("âœ… Model HazÄ±r! CanlÄ± AkÄ±ÅŸ BaÅŸlÄ±yor.")

# 2. JAVASCRIPT: VÄ°DEO AKARKEN ÃœSTÃœNE Ã‡Ä°ZÄ°M YAPMA (Overlay)
def start_overlay_stream():
  js = Javascript('''
    // Global deÄŸiÅŸkenler
    var video;
    var div = null;
    var stream;
    var captureCanvas;
    var overlayCanvas;
    var overlayCtx;

    var pendingResolve = null;
    var shutdown = false;

    function removeDom() {
       stream.getVideoTracks()[0].stop();
       video.remove();
       div.remove();
       video = null;
       div = null;
       stream = null;
    }

    async function createDom() {
      if (div !== null) {
        return stream;
      }

      div = document.createElement('div');
      div.style.border = '2px solid black';
      div.style.padding = '3px';
      div.style.width = '100%';
      div.style.maxWidth = '640px';
      div.style.position = 'relative'; // Overlay iÃ§in relative pozisyon
      document.body.appendChild(div);

      // Durum YazÄ±sÄ±
      const modelOut = document.createElement('h3');
      modelOut.innerText = "CANLI VÄ°DEO (YOLOv5)";
      modelOut.style.color = "green";
      div.appendChild(modelOut);

      // Video Elementi
      video = document.createElement('video');
      video.style.display = 'block';
      video.width = 640;
      video.height = 480;
      video.setAttribute('playsinline', '');

      // AkÄ±ÅŸÄ± Al
      stream = await navigator.mediaDevices.getUserMedia({video: { facingMode: "environment"}});
      video.srcObject = stream;
      await video.play();

      // Video div'e ekle
      div.appendChild(video);

      // Ã‡izim KatmanÄ± (Canvas) - Videonun tam Ã¼stÃ¼ne oturacak
      overlayCanvas = document.createElement('canvas');
      overlayCanvas.width = 640;
      overlayCanvas.height = 480;
      overlayCanvas.style.position = 'absolute'; // Videonun Ã¼stÃ¼ne binmesi iÃ§in
      overlayCanvas.style.left = '3px'; // div padding kadar kaydÄ±r
      overlayCanvas.style.top = modelOut.offsetHeight + 3 + 'px'; // BaÅŸlÄ±k yÃ¼ksekliÄŸi + padding
      overlayCanvas.style.zIndex = '10'; // En Ã¼stte gÃ¶rÃ¼nsÃ¼n

      overlayCtx = overlayCanvas.getContext('2d');
      div.appendChild(overlayCanvas);

      // GÃ¶rÃ¼nmez canvas (Resmi yakalayÄ±p Python'a atmak iÃ§in)
      captureCanvas = document.createElement('canvas');
      captureCanvas.width = 640;
      captureCanvas.height = 480;

      return stream;
    }

    // Python'dan gelen kutularÄ± Ã§izen fonksiyon
    window.drawBoxes = function(boxes) {
      if (!overlayCtx) return;

      // Eski Ã§izimleri temizle
      overlayCtx.clearRect(0, 0, 640, 480);

      // Yeni kutularÄ± Ã§iz
      boxes.forEach(box => {
        // box = [x1, y1, x2, y2, label, color]
        const x = box[0];
        const y = box[1];
        const w = box[2] - box[0];
        const h = box[3] - box[1];
        const label = box[4];

        // Kutu
        overlayCtx.beginPath();
        overlayCtx.lineWidth = 3;
        overlayCtx.strokeStyle = "#00FF00"; // YeÅŸil Ã‡erÃ§eve
        overlayCtx.rect(x, y, w, h);
        overlayCtx.stroke();

        // YazÄ± Arka PlanÄ±
        overlayCtx.fillStyle = "#00FF00";
        overlayCtx.fillRect(x, y - 20, w, 20);

        // YazÄ±
        overlayCtx.fillStyle = "black";
        overlayCtx.font = "bold 16px Arial";
        overlayCtx.fillText(label, x + 5, y - 5);
      });
    }

    // Python'a kare gÃ¶nderen fonksiyon
    async function stream_frame() {
      if (shutdown) {
        removeDom();
        shutdown = false;
        return '';
      }

      await createDom();

      // Videodan kare al
      captureCanvas.getContext('2d').drawImage(video, 0, 0, 640, 480);
      var result = captureCanvas.toDataURL('image/jpeg', 0.5); // HÄ±z iÃ§in kalite dÃ¼ÅŸÃ¼k

      return result;
    }
    ''')
  display(js)

def video_frame():
  data = eval_js('stream_frame()')
  return data

def update_boxes(boxes_json):
  # Python listesini JS tarafÄ±na gÃ¶nderip Ã§izdiren komut
  # boxes_json formatÄ±: [[x1, y1, x2, y2, "Label"], ...]
  eval_js(f'window.drawBoxes({boxes_json})')

def js_to_image(js_reply):
    image_bytes = base64.b64decode(js_reply.split(',')[1])
    jpg_as_np = np.frombuffer(image_bytes, dtype=np.uint8)
    img = cv2.imdecode(jpg_as_np, flags=1)
    return img

# 3. ANA DÃ–NGÃœ
print("Kamera baÅŸlatÄ±lÄ±yor...")
start_overlay_stream()

while True:
    try:
        # 1. GÃ¶rÃ¼ntÃ¼yÃ¼ JS'den iste
        js_reply = video_frame()
        if not js_reply: break

        # 2. Ä°ÅŸlenebilir formata Ã§evir
        img = js_to_image(js_reply)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # 3. Model Tahmini
        results = model(img_rgb)

        # 4. KoordinatlarÄ± Ã‡Ä±kar (Resim Ã§izmek yerine koordinat topluyoruz)
        boxes_to_send = []
        for *xyxy, conf, cls in results.xyxy[0]:
            x1, y1, x2, y2 = int(xyxy[0]), int(xyxy[1]), int(xyxy[2]), int(xyxy[3])
            label = f'{model.names[int(cls)]} {conf:.2f}'

            # JS'ye gÃ¶nderilecek listeye ekle
            boxes_to_send.append([x1, y1, x2, y2, label])

        # 5. KoordinatlarÄ± JS'ye gÃ¶nder (Video arka planda akmaya devam ederken kutular gÃ¼ncellenir)
        update_boxes(str(boxes_to_send))

    except KeyboardInterrupt:
        break
    except Exception as e:
        # Hata olsa bile devam et, akÄ±ÅŸÄ± bozma
        pass
````

````python
def manuel_karne_hesapla(precision, recall, map_score):
    # F1 Skoru FormÃ¼lÃ¼: 2 * (P * R) / (P + R)
    f1_score = 2 * (precision * recall) / (precision + recall)

    print("="*40)
    print(f"ðŸ“Š MODEL KARNESÄ° (Sizin Verileriniz)")
    print("="*40)
    print(f"âœ… Kesinlik (Precision): {precision}  (YanlÄ±ÅŸ alarm oranÄ± dÃ¼ÅŸÃ¼k)")
    print(f"âœ… DuyarlÄ±lÄ±k (Recall):  {recall}  (Nesnelerin %{int(recall*100)}'sini yakalÄ±yor)")
    print(f"âœ… mAP@0.5 (Genel Puan): {map_score}  (Model kalitesi)")
    print("-" * 40)
    print(f"ðŸ† F1 SKORU:           {f1_score:.4f}")
    print("="*40)

    print("\nðŸ“ YORUM:")
    if f1_score > 0.8:
        print("MÃœKEMMEL SONUÃ‡! Model neredeyse hatasÄ±z.")
    elif f1_score > 0.6:
        print("Ä°YÄ° SONUÃ‡. Model Ã§oÄŸu nesneyi tanÄ±yor, canlÄ± kullanÄ±m iÃ§in gayet uygun.")
    else:
        print("ORTA/ZAYIF SONUÃ‡. Daha fazla veriye ihtiyaÃ§ var.")


# P = 0.705, R = 0.62, mAP = 0.704
manuel_karne_hesapla(0.705, 0.62, 0.704)
````
